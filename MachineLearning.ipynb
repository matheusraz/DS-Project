```python
from IPython.display import display

import numpy as np
import pandas as pd
from sklearn.neural_network import MLPClassifier
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('vgsalesPP.csv')
df.drop(['Unnamed: 0'],axis=1,inplace=True)
df
```

```python
df.corr()
```

```python
df_LR = df[['NA_Sales','EU_Sales']].copy()
classes_LR = df['Other_Sales'].copy()
display(df_LR)
classes_LR
```

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(df_LR,classes_LR,test_size = 0.3)

model = LinearRegression()
for i in range(100):
    model.fit(x_treino, y_treino)
score_regressao = model.score(x_teste, y_teste)
print("Porcentagem de acerto da regress√£o linear: {}%".format(score_regressao*100))
```

```python
display(list(df.columns).index('Action'))
display(list(df.columns).index('Strategy'))
print('-----------------------')
display(list(df.columns).index('AO'))
display(list(df.columns).index('T'))
df.columns
```

```python
first = df.loc[:,'Action':'Strategy'].copy()
second = df.loc[:,'AO':'T'].copy()
df_RNA = pd.concat([first.reset_index(drop=True),second.reset_index(drop=True)],axis=1)
classes_RNA = df['Global_Sales'].copy()
display(df_RNA)
classes_RNA
```

```python
df.corr()
```

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(df_RNA,classes_RNA,test_size = 0.3)

mlp = MLPRegressor(hidden_layer_sizes=(3), 
                  activation='tanh', solver='lbfgs')
for i in range(100):
    mlp.fit(x_treino, y_treino)
    
mlp.predict(x_teste)
result = mlp.score(x_teste,y_teste)
print("{}%".format(result*100))
```
